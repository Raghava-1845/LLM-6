```python
# Import CountVectorizer from sklearn
from sklearn.feature_extraction.text import CountVectorizer
 # Take single sentence input (changed to previous text)
sentence = "My name is raghava and I am a student"
 # Initialize the CountVectorizer
vectorizer = CountVectorizer()
 # Generate Bag of Words (BoW) embedding
# CountVectorizer expects input as a list of sentences
bow_embedding = vectorizer.fit_transform([sentence])
 # Print the vocabulary created by the vectorizer
print("Vocabulary:",
vectorizer.get_feature_names_out())
# Print the BoW embedding as an array
print("BoW Embedding:", bow_embedding.toarray())
```

```python
# Install gensim (run once if not installed)
# !pip install gensim
 # Import Word2Vec from gensim
from gensim.models import Word2Vec
 # Input sentence (changed to previous text)
sentence = "My name is raghava and Iam a student"
 # Tokenize the sentence (convert to lowercase and split into words)
tokenized_sentence = sentence.lower().split()
 # Train the Word2Vec model
model = Word2Vec(
 sentences=[tokenized_sentence], # Training data
 vector_size=100, # Size of word vectors
 window=5, # Context window size
 min_count=1, # Include all words
 workers=4 # Number of CPU threads
)
 # Display word embeddings
print("\nWord Embeddings:")
for word in tokenized_sentence:
 print(f"{word} : {model.wv[word]}")
```

```python
# Import gensim downloader to load pretrained embeddings
import gensim.downloader as api

# Load pretrained GloVe model (50-dimensional vectors)
glove_model = api.load("glove-wiki-gigaword-50")

# Input sentence
sentence = "My name is raghava and Iam a student"

# Tokenize the sentence (lowercase and split into words)
words = sentence.lower().split()

# Display GloVe embeddings
print("\nGloVe Word Embeddings:")
for word in words:
    if word in glove_model:
        print(f"{word} : {glove_model[word]}")
    else:
        print(f"{word} : Not found in GloVe vocabulary")
```

```python
# Install required libraries (run once if not installed)
# !pip install transformers torch
 # Import required libraries
import torch
from transformers import BertTokenizer, BertModel
 # Load pretrained BERT tokenizer and model
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model = BertModel.from_pretrained("bert-base-uncased")
# Input sentence (changed to previous text)
sentence = "My name is raghava and Iam a student"
 # Tokenize the input sentence
inputs = tokenizer(
 sentence,
 return_tensors="pt",
 padding=True,
 truncation=True
)
 # Generate embeddings without gradient computation
with torch.no_grad():
 outputs = model(**inputs)
 # Get the last hidden state
# Shape: (batch_size, sequence_length, hidden_size)
last_hidden_state = outputs.last_hidden_state
 # Generate sentence embedding using mean pooling
sentence_embedding = torch.mean(last_hidden_state, dim=1)
 # Print the embedding shape and vector
print("\nSentence Embedding Shape:", sentence_embedding.shape)
print("Sentence Embedding Vector:\n", sentence_embedding)
```

```python
# Take input sentence from user
sentence = input("Enter a sentence: ")

# Tokenize sentence
words = sentence.lower().split()

# Create vocabulary
vocab = sorted(set(words))

# Create one-hot encoding
one_hot_vectors = {}
for word in vocab:
    vector = [1 if w == word else 0 for w in vocab]
    one_hot_vectors[word] = vector

# Display results
print("\nVocabulary:", vocab)
print("\nOne-Hot Encodings:")
for word, vector in one_hot_vectors.items():
    print(f"{word} : {vector}")
```
